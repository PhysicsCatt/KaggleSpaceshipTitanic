First I'll load my datasets

```python
train_data = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
train_data.head(15)
```

```python
test_data = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
test_data.head(15)
```

```python
Earth = train_data.loc[train_data.HomePlanet == 'Earth']["Transported"]
rate_Earth = sum(Earth)/len(Earth)

print("% of earthlings who survived:", rate_Earth)

Europa = train_data.loc[train_data.HomePlanet == 'Europa']["Transported"]
rate_Europa = sum(Europa)/len(Europa)

print("% of Europalings who survived:", rate_Europa)
```

I can see that origin planet has an affect on whether the passenger has transported or not. I'll take this feature into account.

```python
CryoSleep = train_data.loc[train_data.CryoSleep == True]["Transported"]
rate_CryoSleep = sum(CryoSleep)/len(CryoSleep)

print("% of CryoSleepers who survived:", rate_CryoSleep)

CryoSleep = train_data.loc[train_data.CryoSleep == False]["Transported"]
rate_CryoSleep = sum(CryoSleep)/len(CryoSleep)

print("% of Nonsleepers who survived:", rate_CryoSleep)
```

I observe a drastic difference betweem cryosleepers

```python
VIP = train_data.loc[train_data.VIP == True]["Transported"]
rate_VIP = sum(VIP)/len(VIP)

print("% of VIPs who survived:", rate_VIP)

noVIP = train_data.loc[train_data.VIP == False]["Transported"]
rate_noVIP = sum(noVIP)/len(noVIP)

print("% of NoVip who survived:", rate_noVIP)
```

```python
cut = pd.DataFrame(train_data)
mean_value_age_train=cut['Age'].mean()
print(mean_value_age_train)
Young = train_data.loc[train_data.Age <= mean_value_age_train]["Transported"]
rate_Young = sum(Young)/len(Young)
print("% of young people who survived:", rate_Young)


mean_value_age_train=cut['Age'].mean()
Old = train_data.loc[train_data.Age >= mean_value_age_train]["Transported"]
rate_Old = sum(Old)/len(Old)
print("% of old people who survived:", rate_Old)
```

I don't observer a signicant difference below and above average ages. I won't take into account. I want to assess the difference between passengers staying in port and starboard

```python
idf = pd.DataFrame(train_data)
idf['Cabin'].fillna("Unknown", inplace=True)
train_data=idf
```

```python
idf = pd.DataFrame(test_data)
idf['Cabin'] = idf['Cabin'].replace([True],'Port')
test_data=idf
test_data.head(20)
```

```python
idf = pd.DataFrame(train_data)
idf['Cabin']=idf['Cabin'].replace('S', True, regex=True)
train_data=idf
train_data.head(20)
```

```python
idf = pd.DataFrame(train_data)
idf['Cabin'] = idf['Cabin'].replace([True],'Star')
train_data=idf
train_data.head(20)
```

```python
idf = pd.DataFrame(test_data)
idf['Cabin'] = idf['Cabin'].replace([True],'Star')
test_data=idf
test_data.head(20)
```

```python
idf = pd.DataFrame(train_data)
idf['CryoSleep'] = idf['CryoSleep'].replace([True],'Sleep')
idf['CryoSleep'] = idf['CryoSleep'].replace([False],'Awake')

idf['VIP'] = idf['VIP'].replace([True],'vip')
idf['VIP'] = idf['VIP'].replace([False],'novip')

train_data=idf
```

```python
idf = pd.DataFrame(test_data)
idf['CryoSleep'] = idf['CryoSleep'].replace([True],'Sleep')
idf['CryoSleep'] = idf['CryoSleep'].replace([False],'Awake')

idf['VIP'] = idf['VIP'].replace([True],'vip')
idf['VIP'] = idf['VIP'].replace([False],'novip')

test_data=idf
```

```python
idf = pd.DataFrame(train_data)
mean_value_age_train=idf['Age'].mean()
idf['Age'].fillna(value=mean_value_age_train, inplace=True)
train_data=idf

idf = pd.DataFrame(test_data)
mean_value_age_test=idf['Age'].mean()
idf['Age'].fillna(value=mean_value_age_test, inplace=True)
test_data=idf
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Transported"]

features = ["HomePlanet", "CryoSleep","Cabin","Destination","Age", "VIP"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)
#print(type(predictions))

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Transported': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
